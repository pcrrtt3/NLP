<b>From midterm report:</b> We will use word2vec embeddings to define an embedding ”axis” that describes a valence spectrum with which we calculate the cosine similarity of sentences containing references to the targets. We will use the 20 words with the highest and lowest valence from the NRC VAD to define the ends of the spectrum.

```python
import spacy
from collections import defaultdict, Counter
import pandas as pd
import numpy as np
import neuralcoref

from gensim.models import KeyedVectors
from gensim.scripts.glove2word2vec import glove2word2vec
import numpy.linalg as LA

from scipy.stats import ttest_ind, mannwhitneyu, sem
```

```python
nlp = spacy.load('en_core_web_sm')
coref = neuralcoref.NeuralCoref(nlp.vocab)
nlp.add_pipe(coref, name='neuralcoref')
```

```python
data=pd.read_csv("../../Data/Intermediate Data/final_all_files.csv")

data['base'] = np.where(data['event_type']=='rally', 1, 0) #recode rally vs other into new variable
```

#### Entity identification
define entities using NER

```python
def process_tokens(text, tokenized):
    doc = nlp(text)
    tokenized.append(doc)
    return tokenized
```

```python
#tokenize all the documents
tokenized = []

for i in data['text']:
    process_tokens(i, tokenized)
```

```python
len(tokenized)
```

```python
#get indicies for trump docs only
trumpdocs = list(data[data['candidate']=='trump'].index)
```

```python
def get_people_mentions(doc, min_count=10):
    """ Extract all of the PERSON mentions in a spacy-processed document.
    Returns a dict mapping each unique person name to a list of spacy entity mentions
    Each spacy entity has the following attributes:
    
    * text
    * start position in document (character)
    * end position in document (character)
    * label (NER category)
    
    https://spacy.io/usage/linguistic-features#named-entities
    """
    #people=defaultdict(list)
    people=Counter()
    for entity in doc.ents:
        if entity.label_ == "PERSON":
            #people[entity.text.lstrip().rstrip()].append(entity)
            people[entity.text]+=1
    return people
```

```python
#make a dictionary of people mentions for each text
people_in_texts = []

for i in range(len(tokenized)):
    if i in trumpdocs:
        people_in_texts.append(get_people_mentions(tokenized[i], min_count = 1))
```

```python
#count up all mentions across texts together
all_mentions=Counter()

for dictionary in people_in_texts:
    for person in dictionary.keys():
        all_mentions[person]+=dictionary[person]
```

```python
#count up all mentions across the texts separately for speech and rally

#define a separate dictionary for rally & speech
rally_mentions = Counter()
speech_mentions = Counter()

#add up the mentions in each type
for i in range(len(people_in_texts)):
    if data.loc[data.index[i],'base']==1:
        for person in people_in_texts[i].keys():
            rally_mentions[person]+=people_in_texts[i][person]
    else:
        for person in people_in_texts[i].keys():
            speech_mentions[person]+=people_in_texts[i][person]        
```

#### SemAxis
Get the average sentence representations from word embeddings, define the positive and negative axis ends using the highest/lowest words, calculate cosine simularity

```python
# First we have to convert the Glove format into w2v format; this creates a new file
glove_file="../../Data/Source Data/glove.6B.100d.100K.txt"
glove_in_w2v_format="../../Data/Source Data/glove.6B.100d.100K.w2v.txt"
_ = glove2word2vec(glove_file, glove_in_w2v_format)
```

```python
glove = KeyedVectors.load_word2vec_format(glove_in_w2v_format, binary=False)
```

```python
#define the initial pole words
pole_pos = glove['good']
pole_neg = glove['bad']
```

```python
#use the strategy in An et al section 3.3.2 to expand the initial pole words
#i.e., find the 20 highest and lowest words, compute geometric center to define the axis

#step 1: find the l closest words to the initial word
pos_sims=glove.most_similar(positive=pole_pos,topn=21) #using top 21 bc it includes the initial word
neg_sims=glove.most_similar(positive=pole_neg,topn=21)

print('Positive pole words')
for x in pos_sims:
    print(x[0],x[1])

print('')

print('Negative pole words')
for x in neg_sims:
    print(x[0],x[1])
```

```python
#step 2: compute the geometric center of l+1 words - this defines each pole - then construct axis
def get_semaxis(vectors, positive_terms, negative_terms):
    
    pos_vecs = [vectors[positive_terms[i][0]] for i in range(len(positive_terms))]

    neg_vecs = [vectors[negative_terms[i][0]] for i in range(len(negative_terms))]

    v_plus=np.mean(pos_vecs, axis=0)
    v_neg=np.mean(neg_vecs, axis=0)
    
    v_axis=v_plus-v_neg
    
    return v_axis

axis = get_semaxis(glove, pos_sims, neg_sims) #this is the axis we'll use to get a score
```

```python
#functions that will calculate the geometric centers for all words in the sentences in which our top 10 entities are found

def embedding_func(token):
    # helper function that takes a token and returns word embedding or passes if word is not in glove
    try:
        return glove[token.text.lower()]
    except:
        pass


def entity_sentence_embeddings(text):
    #input: tokenized text
    #output: a list of tuples (token, average embedding for sentence that contains a person entity in our top 10)
    sentences = []
    
    for token in text:
        if token.ent_type_=="PERSON" and token.text in top10:
            
            #make list of tokens in sentence
            words = [word for word in token.sent]
            
            #get word embeddings for each word and take average of sentence
            try:
                sentences.append( (token, np.mean([embedding_func(word) for word in words], axis=0) ) )
            except:
                pass 
            
    return( sentences )


def score_list_of_targets(vectors, sem_axis, target_vecs):
    scores=[]
    for entity, array in target_vecs:
        scores.append( (entity, vectors.cosine_similarities(array, [sem_axis])[0]) )
    return(scores)
```

```python
#two dictionaries of the top ten - key is name, value is list of scores
rally_dict = {}
speech_dict = {}

#split tokenized text data into rally and speech using indices from data frame
rallies = [tokenized[i] for i in list(data.index[data['base']==1]) if i in trumpdocs] 
speeches = [tokenized[i] for i in list(data.index[data['base']==0]) if i in trumpdocs] 

def scores_to_dict(tokenized_texts, score_dict):
    for i in tokenized_texts: #iterate thru each tokenized text
    
        embeds = entity_sentence_embeddings(i) #get sentence embeddings for top 10

        scores = score_list_of_targets(glove, axis, embeds) #get cosine similarity scores for each sentence
    
        for entity, score in scores: #loop thru list of scores and add to the appropriate dictionary entity entry
            if entity.text in score_dict.keys():
                score_dict[entity.text].append(score)
            else:
                score_dict[entity.text]=[score]
```

```python
scores_to_dict(rallies, rally_dict)
```

```python
scores_to_dict(speeches, speech_dict)
```

```python
#Welch's t-tests - for unequal sample sizes with unequal variance

for i in top10:
    try:
        noperm = ttest_ind(rally_dict[i], speech_dict[i], equal_var=False)
        perm = ttest_ind(rally_dict[i], speech_dict[i], equal_var=False, permutations = 100)
        print(i)
        print('No permutation:', noperm)
        print('N rally =', len(rally_dict[i]), 'N speech =', len(speech_dict[i]))
        print('With permutation:', perm)
        print('')
    except:
        pass
```

```python
#Mann-whitney tests - replicating results of t-tests for robustness check

for i in top10:
    try:
        print(i)
        print(mannwhitneyu(rally_dict[i], speech_dict[i]))
        print('N rally =', len(rally_dict[i]), 'N speech =', len(speech_dict[i]))
        print('')
    except:
        pass
```
