```python
import spacy
from collections import defaultdict
import networkx as nx
import matplotlib.pyplot as plt
from nltk.tokenize import sent_tokenize, word_tokenize
import pandas as pd
import numpy as np
import re
import os
```

```python
nlp = spacy.load('en_core_web_sm') #, disable=['parser']
```

```python
inpath2= os.path.dirname(os.path.dirname(os.getcwd()))+"\\Data\\Intermediate Data\\"
df=pd.read_csv(inpath2+'final_all_files.csv')
```

```python
df=df[df['candidate']=='trump']
df['rally']=np.where(df['event_type']=='rally', 1, 0)
```

```python
text=df['text'].to_list()
rally=df['rally'].to_list()
title=df['title'].to_list()
```

```python
def get_people_mentions(data, min_count=1):
    doc=nlp(data)
    people=defaultdict(list)
    for entity in doc.ents:
        if entity.label_ == "PERSON":
            people[entity.text.lstrip().rstrip()].append(entity)
    
    return people
```

```python
def get_nodes(people, min_count=1, title=' '):
    nodes=defaultdict(float)
    for person in people:
        if len(people[person]) >= min_count:
            nodes[person]=len(people[person])
    nodes['title']=title
    return nodes
```

```python
people=get_people_mentions(text[0])
nodes=get_nodes(people)
```

```python
all_people=[]
for spe, title in zip(text, title):
    print(title)
    nodes=get_nodes(get_people_mentions(spe), title=title)
    all_people.append(nodes)
```

```python
top_people=pd.DataFrame(all_people)
```

```python
top_people_freq=top_people.copy()
del top_people_freq['title']
top_people_freq=top_people_freq.sum().reset_index()
top_people_freq.rename(columns={0:'frequency'}, inplace=True)
```

```python
top_people_freq.sort_values(by='frequency', ascending=False, inplace=True)
```

```python
biden=re.compile(r'nancy|pelosi', re.IGNORECASE)

biden_list=['Biden', 'Joe', 'Joe Biden', 'BIDEN', "Joe Biden's", 'Slow Joe', "JOE BIDEN'S"]
clinton_list=['Hillary', 'Hillary Clinton', 'Clinton', 'Bill Clinton',
       'Crooked Hillary', "Hillary Clinton's",
       'DONNA BRAZILE GET CROOKED HILLARY', 'HILLARY', 'HILLARY CLINTON',
       'Clintons', 'Hillary Madam', 'Hillary Rodham Clinton']
obama_list=['Obama', 'Barack Obama', 'Barack Hussein Obama', 'Michelle Obama',
       'Barack-Hussein Obama', 'Obama-Biden', 'Barack-Hussein-Obama',
       'Obama’s', "Obama-Biden's", "Barack Obama's", 'Impeach Obama',
       'Barack', 'Obama-Biden-Santos', 'Barak Obama', 'Barack H. Obama']
pelosi_list=['Nancy Pelosi', 'Nancy', 'Nervous Nancy', "Nancy Pelosi's",
       'Nancy Polosi', 'Nancy Pelosi’s', 'Nancy Pelosi-controlled',
       'Chuck-and-Nancy', 'Pelosi']
```

```python
top_people_freq[top_people_freq['index'].str.contains(biden)]['index'].unique()
```

```python
top_people['biden_references']=top_people[biden_list].sum(axis=1)
top_people['clinton_references']=top_people[clinton_list].sum(axis=1)
top_people['obama_references']=top_people[obama_list].sum(axis=1)
top_people['pelosi_references']=top_people[pelosi_list].sum(axis=1)
```

```python
top_people2=pd.merge(top_people[['title', 'biden_references', 'clinton_references', 'obama_references', 'pelosi_references' ]], df[['title', 'rally']], on=['title'], how='left')
```

```python
for i in ['biden_references','clinton_references', 'obama_references', 'pelosi_references']:
    top_people2["{}_bin".format(i)]=np.where(top_people2[i]>0, 1, 0)
```

```python
top_people2.groupby('rally').mean()
```

```python
def get_norp_mentions(data, min_count=1):
    doc=nlp(data)
    norp=defaultdict(list)
    for entity in doc.ents:
        if entity.label_ == "NORP":
            norp[entity.text.lstrip().rstrip()].append(entity)
    
    return norp
```

```python
text=df['text'].to_list()
rally=df['rally'].to_list()
title=df['title'].to_list()
all_norp=[]
for spe, title in zip(text, title):
    print(title)
    nodes=get_nodes(get_norp_mentions(spe), title=title)
    all_norp.append(nodes)
```

```python
top_norp=pd.DataFrame(all_norp)
```

```python
top_norp_freq=top_norp.copy()
del top_norp_freq['title']
top_norp_freq=top_norp_freq.sum().reset_index()
top_norp_freq.rename(columns={0:'frequency'}, inplace=True)
top_norp_freq.sort_values(by='frequency', ascending=False, inplace=True)
```

```python
democrat=re.compile(r'demo|lib|soc|left', re.IGNORECASE)
top_norp_freq[top_norp_freq['index'].str.contains(democrat)]['index'].unique()
```

```python
democrat_list=['Democrats', 'Democrat', 'democrats', 'democrat',
       'DEMOCRATS', 'DEMOCRAT', 'Liberals', 'Socialist', 
       'Socialists', 'Left Democrats',
       'Socialist Democrats']
top_norp['democrat_references']=top_norp[democrat_list].sum(axis=1)
top_norp2=pd.merge(top_norp[['title', 'democrat_references' ]], df[['title', 'rally']], on=['title'], how='left')
for i in ['democrat_references']:
    top_norp2["{}_bin".format(i)]=np.where(top_norp2[i]>0, 1, 0)
top_norp2.groupby('rally').mean()
```
